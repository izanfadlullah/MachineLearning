```python
# Supervised Machine Learning
# Regression - Polynomial Regression

# Import ALL neccessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
lin_reg = lm.LinearRegression()
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv')
data
```

```python
# Select Dependent and Independent Variable
X = data.Level.values.reshape(-1,1)
Y = data.Salary.values
```

```python
# Model Development
lin_reg.fit(X,Y)
```

```python
pred_values = lin_reg.predict(X)
print(pred_values)
```

```python
plt.scatter(X,Y,color='red',label='Actual Salaries')
plt.title('Salaries vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.plot(X,pred_values,color='blue')
plt.show()
```

```python
# Use Normalization
import sklearn.preprocessing as pp
y_scaled = pp.normalize([Y]).flatten()
y_scaled
```

```python
y_scaled.shape
```

```python
# Model Development after Normalization
lin_reg.fit(X,y_scaled)
```

```python
plt.scatter(X,y_scaled,color='red',label='Actual Salaries')
plt.title('Salaries vs Level after Normalization')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.show()
```

```python
# Let's Use Polynomial Regression
# Generate Polynomial Features

# by default the degree of independent variable is 1
# but in Polynomial regression we will increase the degree of independent variable
# and then we will Train Our Model

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)

x_poly_2 = poly_reg.fit_transform(X)
x_poly_2

# when degree = 2 , then new features will be created like this  1  x*1  x*x
# when degree = 3 , then new features will be created like this  1  x*1  x*x  x*x*x
# when degree = 4 , then new features will be created like this  1  x*1  x*x  x*x*x  x*x*x*x
```

```python
# Model Development Using Degree = 2
lin_reg_2 = lm.LinearRegression()
lin_reg_2.fit(x_poly_2,Y)
```

```python
# plot actual values
plt.scatter(X, Y, color = 'red', label='Actual Values')
# plot the regression line
plt.plot(X,lin_reg_2.predict(x_poly_2), color = 'green', label='Regression Line')
plt.title('Regression Line After Using Degree = 2')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
# Degree = 3
poly_reg = PolynomialFeatures(degree=3)
x_poly_3 = poly_reg.fit_transform(X)
x_poly_3
# Model Development Using Degree = 3
lin_reg_3 = lm.LinearRegression()
lin_reg_3.fit(x_poly_3,Y)
```

```python
# plot actual values
plt.scatter(X, Y, color = 'red', label='Actual Values')
# plot the regression line
plt.plot(X,lin_reg_3.predict(x_poly_3), color = 'green', label='Regression Line')
plt.title('Regression Line After Using Degree = ')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
# Degree = 4
poly_reg = PolynomialFeatures(degree=4)
x_poly_4 = poly_reg.fit_transform(X)
x_poly_4
# Model Development Using Degree = 4
lin_reg_4 = lm.LinearRegression()
lin_reg_4.fit(x_poly_4,Y)
```

```python
# plot actual values
plt.scatter(X, Y, color = 'red', label='Actual Values')
# plot the regression line
plt.plot(X,lin_reg_4.predict(x_poly_4), color = 'green', label='Regression Line')
plt.title('Regression Line After Using Degree = 4 ')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
# using for loop to make PolynomialFeatures
for i in range(1,5):
  poly_reg = PolynomialFeatures(degree=i)
  x_poly = poly_reg.fit_transform(X)
  x_poly
  lin_reg = lm.LinearRegression()
  lin_reg.fit(x_poly,Y)
```

```python
# plot actual values
plt.scatter(X, Y, color = 'red', label='Actual Values')
# plot the regression line
plt.plot(X,lin_reg.predict(x_poly), color = 'green', label='Regression Line')
plt.title('Regression Line After Using Degree')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
# plot actual values
plt.scatter(X,Y,color='red',label='Actual Values')
for i in range(2,5):
  poly_reg = PolynomialFeatures(degree=i)
  x_poly = poly_reg.fit_transform(X)
  lin_reg.fit(x_poly,Y)
  plt.plot(X,lin_reg.predict(x_poly),label=f'Degree{i}')

plt.title('Regression Line After Using Different Degrees')
plt.xlabel('Level')
plt.ylabel('Salaries')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(Y,lin_reg.predict(x_poly))
```

```python
# Final Code For Polynomial Regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv')
X = data[['Level']].values
y = data['Salary'].values

# Try polynomial degrees 1 to 5 (no X_grid)
plt.scatter(X, y, color='red', label='Actual Data')

for degree in range(1, 6):
    # Create polynomial features
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)

    # Fit linear regression
    lin_reg = LinearRegression()
    lin_reg.fit(X_poly, y)

    # Predict using the same X (no grid)
    y_pred = lin_reg.predict(X_poly)

    # Calculate r2 score
    r2 = r2_score(y, y_pred)
    print(f"Degree {degree} - R2 Score: {r2}")

    # Plot the curve
    plt.plot(X, y_pred, label=f'Degree {degree}')

plt.title("Polynomial Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
